### Foreword
Since I majored in math and extensively worked through principal component analysis (PCA) and factor analysis (FA) during my master's degree from a mathematical standpoint, my goal here is to heuristically connect the two approaches to an imaginary use case in drug discovery and illustrate their difference. I will start with PCA and then move on to FA.


### Scenario

Assume a 10-dimensional drug profile ($x$) meaning a single compound is represented by a 10-dimensional vector with each dimension describing a physical or biological measurement - also called feature - e.g. mutagenicity, lethal dose,...
and without knowing which ones are important we want to extract the 2 "most important" features from the space. 
Let the example vector with 10 dimensions be:
$$ X = 
\begin{pmatrix} 
\dots & x_1 & \dots \\
 & \vdots &  \\
 \dots & x_{1000} & \dots 
\end{pmatrix} \in \mathbb{R}^{1000 \times 10}$$
where $x_i \in \mathbb{R}^{10}$ could for example be of the form:
$$x_i= [500, 4.2, 85, 0.1, 0.2, 95, 10, 12, 0.8, 7.5]$$
and the 10 dimensions describe the following features:
| Index | Feature | Meaning | Unit / Scale |
| :--- | :--- | :--- | :--- |
| 1 | **MW** | Molecular Weight | Daltons |
| 2 | **LogP** | Lipophilicity | Octanol-water partition |
| 3 | **HBD** | Hydrogen Bond Donors | Integer Count |
| 4 | **hERG** | Heart Toxicity | % Inhibition |
| 5 | **Ames** | Mutagenicity | Probability (0 to 1) |
| 6 | **PSA** | Polar Surface Area | $\text{\AA}^2$ |
| 7 | **LD50** | Lethal Dose 50 | mg/kg |
| 8 | **CYP3A4** | Liver Metabolism | Clearance ($L/hr$) |
| 9 | **Solubility**| Water Solubility | mg/mL |
| 10 | **Perm** | Cell Permeability | $10^{-6} \text{ cm/s}$ |

### PCA: Change of Basis and Dimensionality Reduction
PCA performs a change of basis on the data and then reduces dimensionality. Mathematically, PCA does not "delete" anything initially. It performs a rotation of the coordinate system. We find an orthonormal basis $V$ (the Eigenvectors of either the covariance matrix or the correlation matrix) and transform $x$ into $z$:

$$z = x V $$

At this stage, **$z$ is still 10-dimensional**. We have simply changed our perspective to align with the directions of maximum variance in the dataset. Only if we choose a submatrix of $V$ (or a subvector of $z$) do we really reduce dimensions.

#### 1. Mathematical Transformation: The Change of Basis
Consider the 1000 different compounds. To perform PCA, we first compute the covariance or correlation matrix. The $10 \times 10$ symmetric covariance matrix is defined by :
$$
\Sigma = \left(Cov(x_i,x_j)\right)_{i,j=1}^{10}
=
\frac{1}{n-1}X^\top X
$$
and could have the form (heuristic example):
$$
\Sigma
=
\begin{pmatrix}
10000.0 & 98.4 & 130.0 & 0.75 & 0.20 & 1125.0 & 2.5 & 32.0 & -12.0 & 71.0 \\
98.4 & 1.44 & 1.32 & 0.005 & 0.001 & 11.7 & 0.006 & 0.02 & -0.04 & 0.96 \\
130.0 & 1.32 & 4.00 & 0.003 & 0.008 & 24.0 & 0.01 & 0.02 & -0.16 & 1.00 \\
0.75 & 0.005 & 0.003 & 0.023 & 0.013 & 0.007 & 1.57 & 0.21 & 0.001 & 0.002 \\
0.20 & 0.001 & 0.008 & 0.013 & 0.010 & 0.005 & 0.66 & 0.09 & 0.001 & 0.001 \\
1125.0 & 11.7 & 24.0 & 0.007 & 0.005 & 225.0 & 0.03 & 0.24 & -1.50 & 10.2 \\
2.5 & 0.006 & 0.01 & 1.57 & 0.66 & 0.03 & 144.0 & 16.8 & 0.03 & 0.05 \\
32.0 & 0.02 & 0.02 & 0.21 & 0.09 & 0.24 & 16.8 & 16.0 & 0.07 & 0.20 \\
-12.0 & -0.04 & -0.16 & 0.001 & 0.001 & -1.50 & 0.03 & 0.07 & 0.04 & -0.12 \\
71.0 & 0.96 & 1.00 & 0.002 & 0.001 & 10.2 & 0.05 & 0.20 & -0.12 & 1.00
\end{pmatrix}
$$
Since the variance is not normalized the choice of units or numerical differences could bias the outcome. Thus we often use the normalized version of the covariance matrix also called the correlation matrix:
$$R = \frac{1}{n-1} X_{norm}^\top X_{norm}$$
where $X_{norm}$ is the normalized data corresponding to $X$ defined by $X_{norm} = (X - \mu)D^{-1}$, $\frac{1}{n-1}$ is the Bessel Correction and  

$\mu = \begin{pmatrix}
\bar{X_1} & \dots & \bar{X_{10}}\\
\vdots & \dots & \vdots \\
\bar{X_1} & \dots & \bar{X_{10}}
 \end{pmatrix} 
$
with $\bar{X_{i}} = \frac{1}{1000}\sum_{j=1}^{1000} X_{j,i} \quad \forall \quad i \in [10]$ and $D = \begin{pmatrix}
\sigma_1 & 0 & \dots \\
\vdots & \ddots & \vdots \\
\dots & 0 & \sigma_{10}
\end{pmatrix} 
$

Then we would get a matrix that would be of the form (heuristic example):
$$
R= \begin{pmatrix}
1.00 & 0.82 & 0.65 & 0.05 & 0.02 & 0.75 & 0.01 & 0.08 & -0.60 & 0.71 \\
0.82 & 1.00 & 0.55 & 0.03 & 0.01 & 0.65 & 0.02 & 0.05 & -0.70 & 0.80 \\
0.65 & 0.55 & 1.00 & 0.01 & 0.04 & 0.60 & 0.03 & 0.02 & -0.40 & 0.50 \\
0.05 & 0.03 & 0.01 & 1.00 & 0.88 & 0.02 & 0.85 & 0.72 & 0.01 & 0.02 \\
0.02 & 0.01 & 0.04 & 0.88 & 1.00 & 0.03 & 0.82 & 0.68 & 0.02 & 0.01 \\
0.75 & 0.65 & 0.60 & 0.02 & 0.03 & 1.00 & 0.01 & 0.04 & -0.50 & 0.68 \\
0.01 & 0.02 & 0.03 & 0.85 & 0.82 & 0.01 & 1.00 & 0.70 & 0.01 & 0.02 \\
0.08 & 0.05 & 0.02 & 0.72 & 0.68 & 0.04 & 0.70 & 1.00 & 0.03 & 0.05 \\
-0.60 & -0.70 & -0.40 & 0.01 & 0.02 & -0.50 & 0.01 & 0.03 & 1.00 & -0.60 \\
0.71 & 0.80 & 0.50 & 0.02 & 0.01 & 0.68 & 0.02 & 0.05 & -0.60 & 1.00
\end{pmatrix}
$$
Afterwards we compute the eigenvalue decomposition $R = V \Lambda V^\top$ of $R$ with
$$V = 
\begin{pmatrix} 
\vdots &  & \vdots \\
v_1 & ... & v_{10} \\
\vdots &  & \vdots 
\end{pmatrix}, \qquad 
\Lambda = 
\begin{pmatrix} 
\lambda_1 &  &  \\
 & \ddots &  \\
 &  & \lambda_{10} 
\end{pmatrix}
$$
where each column $v_i$ is an eigenvector of $R$, that is $R v_i = \lambda_i v_i$ and $\Lambda $ is the matrix with the eigenvalues on the diagonal. Note that $V$ is a full $10 \times 10$ matrix; it only "collapses" if we choose to ignore specific columns. In our example we get:

$$V = \begin{bmatrix} 
\text{PC1 }= v_1 & \text{PC2 }= v_2 & \dots & \text{PC10} =v_{10} \\
\downarrow & \downarrow & & \downarrow \\
0.45 & 0.02 & \dots & 0.01 \\
0.42 & 0.01 & \dots & -0.05 \\
0.38 & 0.05 & \dots & 0.12 \\
0.02 & 0.48 & \dots & -0.02 \\
0.01 & 0.45 & \dots & 0.03 \\
0.40 & 0.03 & \dots & 0.01 \\
0.03 & 0.46 & \dots & -0.10 \\
0.05 & 0.42 & \dots & 0.05 \\
-0.35 & 0.02 & \dots & 0.20 \\
0.41 & 0.01 & \dots & -0.02 
\end{bmatrix}$$

### The Mathematical Properties
1. **Dimensions:** $V$ is $10 \times 10$. It takes a 10D input and provides a 10D output.
2. **PC1 (The "Bioavailability" Basis):** The weights are high for MW (0.45), LogP (0.42), HBD (0.38), PSA (0.40), and Perm (0.41). This axis summarizes the physical movement of the drug because all the bioavailability features vary together.
3. **PC2 (The "Safety" Basis):** The weights are high for hERG (0.48), Ames (0.45), and LD50 (0.46), CYP3A4 (0.42). This axis summarizes the danger level, because all the relevant features vary together.

#### 2. The Collapse (Dimensionality Reduction)

To achieve **collapse**, we perform a projection. We look at the variance which is given by the eigenvalues ($\lambda_1$,...,$\lambda_{10}$) captured by each new axis. In our drug dataset, we find that the first two axes capture almost all the "variance":
To decide how many dimensions to keep, we look at the **Eigenvalues ($\lambda$)** and the percentage of total variance they explain. In a 10-variable system, the sum of all eigenvalues is 10.0.

| Component | Eigenvalue ($\lambda$) | % Variance Explained | Cumulative % |
| :--- | :---: | :---: | :---: |
| **PC1 (Bioavailability)** | **5.85** | **58.5%** | 58.5% |
| **PC2 (Safety)** | **2.65** | **26.5%** | **85.0%** |
| PC3 | 0.45 | 4.5% | 89.5% |
| PC4 | 0.30 | 3.0% | 92.5% |
| PC5 | 0.25 | 2.5% | 95.0% |
| PC6 | 0.15 | 1.5% | 96.5% |
| PC7 | 0.12 | 1.2% | 97.7% |
| PC8 | 0.10 | 1.0% | 98.7% |
| PC9 | 0.08 | 0.8% | 99.5% |
| PC10 | 0.05 | 0.5% | 100.0% |

* **$z_1$ (PC1):** High variance. Represents **"Bioavailability Bulk."**
* **$z_2$ (PC2):** High variance. Represents **"Biological Safety."**
* **$z_3 \dots z_{10}$:** Near-zero variance. These axes represent redundant information or tiny fluctuations. We take only the first two columns $$V_{projection} = \begin{pmatrix} 
\vdots & \vdots \\
v_1 &  v_{2} \\
\vdots & \vdots 
\end{pmatrix} \in \mathbb{R}^{10\times 2}$$ of $V$ and multiply it by our original $1 \times 10$ drug vector $x$ which yields the projected feature vector.

$$\text{z}_{projection} = x  V_{projection}  \in \mathbb{R}^{2}$$

The "collapse" is a conscious choice to simplify the data. We **truncate** the vector $z$ by keeping only the first $k$ dimensions (where $k=2$).

$$z_{\text{projection}} = [z_1, z_2]$$

By discarding $z_3$ to $z_{10}$, we reduce the dimensionality from **10D to 2D** while retaining the core "meaning" of the drug. We are left with a 2D signature that tells a chemist everything they need to know about a molecule's potential success.

#### Cave
PCA makes the implicit assumption that "variance" and "meaning" correlate in a meaningful sense. If variance isn't explaining "meaning" or "causation"  PCA can perform badly for predictive models.


### FA: The Difference
In PCA we tried to capture as much variance as possible in fewer dimensions by rotating and reducing dimensions via
$$\text{z}_{projection} = x  V_{projection}  \in \mathbb{R}^{2}$$
In FA we assume, that part of the variance is caused by measurement errors (e.g. through unpreventable measurement errors in the lab some of the entries in $X$ will probably be off). Therefore we try to find the true latent variables $f$ that, together with some error $\varepsilon$ and the Loading matrix $L$ explains the observation through the relationship $x_i = L f_i + \varepsilon _i$. 
The loading matrix is the matrix defined by 
$$ R= LL^\top + \Psi $$
where $R$ is again the correlation matrix from above, $\Psi$ is the measurement error matrix. Heuristically speaking $L$ contains the part of the correlation $R$ that is explained by the factors and $\Psi$ contains the part of the correlation matrix $R$ that is explained by measure errors (noise). 
#### Computing FA
In practice the $x_i = L f_i + \varepsilon _i$ yields two unknowns ($L$ and $\Psi$) for one equation which is an underdetermined linear system, so we have to work around that. First we compute the reduced correlation matrix $R_r=R−\Psi$ (or $\Sigma_r = \Sigma - \Psi $ in the unnormalized case which makes little sense here) and perform an eigendecomposition $R_r = V_r \Lambda V_r^\top $ (or $\Sigma_r = V_r \Lambda V_r^\top$ for the unnormalized case) where $R$  is as before and $\Psi$ is an educated guess. Then we define $L = V_r^{k} \Lambda^{\frac{1}{2}}$ where $k$ indicates how many columns of the matrix $V_r$ we take. Since for us $k=2$ we use the first two eigenvectors. Using a fixed point iteration we can (optionally) iteratively refine $\Psi$ and $L$ up to where the error is below a certain threshold.

Then the factor matrix is defined by 
$$ F = X \Sigma^{-1} L $$
if not normalized or (in our case, since the units cause large numeric discrepancies in numbers, that don't represent difference in meaning) the largely more sensible normalized version:
$$ F = X_{norm}R^{-1} L $$
where again $X, X_{norm} \in \mathbb{R}^{n\times p}=\mathbb{R}^{1000\times 10}$ as before, $\Sigma, R \in \mathbb{R}^{p\times p}=\mathbb{R}^{10\times 10}$ the covariance matrix and correlation matrix respectively, $L \in \mathbb{R}^{p \times k }=\mathbb{R}^{10\times 2}$. Consequentially $F\in \mathbb{R}^{n\times k} = \mathbb{R}^{1000\times 2}$ contains all the factor row vectors $f_i \in \mathbb{R}^{2}$ in its rows. A heuristic example in our case could look like this:
$$ F =
\begin{pmatrix}
 bioavailability & safety & \\
 1.45 & -0.12 & \rightarrow \text{drug candidate 1}\\
-0.85 & 2.10 & \rightarrow \text{drug candidate 2}\\
\vdots & \vdots \\
0.05 & 0.10 & \rightarrow \text{drug candidate 999}\\
1.10 & 1.9 & \rightarrow \text{drug candidate 1000}
\end{pmatrix} \in \mathbb{R}^{1000 \times 2} $$
To give an example interpretation of this factor score matrix $F$: it shows the first drug has excellent bioavailability (high with 1.45) with very good safety (low mutagenicity and toxicity at -0.12). Drug 2 on the other hand is largely unsafe with high toxicity and mutagenicity (2.10) while presenting low bioavailability (-0.85). 


### Comparing PCA and FA
PCA gives 
$z = x V $ or $Z = X V_k $ with $z_i = x_i V $ in matrix form 
while FA gives 
$f = x R^{-1} L $ or $F = X R^{-1} L$ with $f_i = x R^{-1} L $ in matrix form. Where $z$ and $f$ generally correspond to each other in meaning.
The only heuristic difference between $f_i$ and $z_i$ is that in $f_i$ we stripped away the variance we suspect to be caused by noise/error (by applying $\Psi$) that we left untouched in $z_i$. If noise $\varepsilon$ (e.g. lab errors) are known to be 0, then $z_i = f_i$ and if the noise is small, then $z_i \approx f_i$

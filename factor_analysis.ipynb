### Scenario

Assume a 10-dimensional drug profile ($x$) meaning a single compound is represented by a 10-dimensional vector with each dimension describing a physical or biological measurement - also called feature - e.g. mutagenicity, lethal dose,...
and without knowing which ones are important we want to extract the 2 "most important" features from the space. 
Let the example vector with 10 dimensions be:
$$x = [500, 4.2, 85, 0.1, 0.2, 95, 10, 12, 0.8, 7.5]$$
where the dimensions describe the following features:
| Index | Feature | Meaning | Unit / Scale |
| :--- | :--- | :--- | :--- |
| 1 | **MW** | Molecular Weight | Daltons |
| 2 | **LogP** | Lipophilicity | Octanol-water partition |
| 3 | **HBD** | Hydrogen Bond Donors | Integer Count |
| 4 | **hERG** | Heart Toxicity | % Inhibition |
| 5 | **Ames** | Mutagenicity | Probability (0 to 1) |
| 6 | **PSA** | Polar Surface Area | $\text{\AA}^2$ |
| 7 | **LD50** | Lethal Dose 50 | mg/kg |
| 8 | **CYP3A4** | Liver Metabolism | Clearance ($L/hr$) |
| 9 | **Solubility**| Water Solubility | mg/mL |
| 10 | **Perm** | Cell Permeability | $10^{-6} \text{ cm/s}$ |

### PCA: Change of Basis and Dimensionality Reduction
PCA performs a change of basis on the data and then reduces dimensionality.

#### 1. Mathematical Transformation: The Change of Basis
Mathematically, PCA does not "delete" anything initially. It performs a rotation of the coordinate system. We find an orthonormal basis $V$ (the Eigenvectors of the covariance matrix) and transform $x$ into $z$:

$$z = V^T x$$

At this stage, **$z$ is still 10-dimensional**. We have simply changed our perspective to align with the directions of maximum variance in the dataset.

To perform PCA, we first compute the correlation between all 10 properties. This results in a $10 \times 10$ symmetric covariance matrix:
$$
\Sigma = \left(Cov(x_i,x_j)\right)_{i,j=1}^{10}
=
\begin{pmatrix}
1.00 & 0.82 & 0.65 & 0.05 & 0.02 & 0.75 & 0.01 & 0.08 & -0.60 & 0.71 \\
0.82 & 1.00 & 0.55 & 0.03 & 0.01 & 0.65 & 0.02 & 0.05 & -0.70 & 0.80 \\
0.65 & 0.55 & 1.00 & 0.01 & 0.04 & 0.60 & 0.03 & 0.02 & -0.40 & 0.50 \\
0.05 & 0.03 & 0.01 & 1.00 & 0.88 & 0.02 & 0.85 & 0.72 & 0.01 & 0.02 \\
0.02 & 0.01 & 0.04 & 0.88 & 1.00 & 0.03 & 0.82 & 0.68 & 0.02 & 0.01 \\
0.75 & 0.65 & 0.60 & 0.02 & 0.03 & 1.00 & 0.01 & 0.04 & -0.50 & 0.68 \\
0.01 & 0.02 & 0.03 & 0.85 & 0.82 & 0.01 & 1.00 & 0.70 & 0.01 & 0.02 \\
0.08 & 0.05 & 0.02 & 0.72 & 0.68 & 0.04 & 0.70 & 1.00 & 0.03 & 0.05 \\
-0.60 & -0.70 & -0.40 & 0.01 & 0.02 & -0.50 & 0.01 & 0.03 & 1.00 & -0.60 \\
0.71 & 0.80 & 0.50 & 0.02 & 0.01 & 0.68 & 0.02 & 0.05 & -0.60 & 1.00
\end{pmatrix}
$$
Next we compute 
$$V = 
\begin{pmatrix} 
\vdots & ... & \vdots \\
v_1 & ... & v_{10} \\
\vdots & ... & \vdots 
\end{pmatrix}
$$
where each column $v_i$ is an eigenvector of $\Sigma$, that is $\Sigma v_i = \lambda_i v_i$. Note that $V$ is a full $10 \times 10$ matrix; it only "collapses" if we choose to ignore specific columns. In our example we get:

$$V = \begin{bmatrix} 
\text{PC1 }= v_1 & \text{PC2 }= v_2 & \dots & \text{PC10} =v_{10} \\
\downarrow & \downarrow & & \downarrow \\
0.45 & 0.02 & \dots & 0.01 \\
0.42 & 0.01 & \dots & -0.05 \\
0.38 & 0.05 & \dots & 0.12 \\
0.02 & 0.48 & \dots & -0.02 \\
0.01 & 0.45 & \dots & 0.03 \\
0.40 & 0.03 & \dots & 0.01 \\
0.03 & 0.46 & \dots & -0.10 \\
0.05 & 0.42 & \dots & 0.05 \\
-0.35 & 0.02 & \dots & 0.20 \\
0.41 & 0.01 & \dots & -0.02 
\end{bmatrix}$$

### The Mathematical Properties
1. **Dimensions:** $V$ is $10 \times 10$. It takes a 10D input and provides a 10D output.
2. **PC1 (The "Bioavailability" Basis):** The weights are high for MW (0.45), LogP (0.42), HBD (0.38), PSA (0.40), and Perm (0.41). This axis summarizes the physical movement of the drug because all the bioavailability features vary together.
3. **PC2 (The "Safety" Basis):** The weights are high for hERG (0.48), Ames (0.45), and LD50 (0.46), CYP3A4 (0.42). This axis summarizes the danger level, because all the relevant features vary together.
---

#### 2. The Collapse (Dimensionality Reduction)

To achieve **collapse**, we perform a projection. We look at the variance which is given by the eigenvalues ($\lambda_1$,...,$\lambda_{10}$) captured by each new axis. In our drug dataset, we find that the first two axes capture almost all the "variance":
To decide how many dimensions to keep, we look at the **Eigenvalues ($\lambda$)** and the percentage of total variance they explain. In a 10-variable system, the sum of all eigenvalues is 10.0.

| Component | Eigenvalue ($\lambda$) | % Variance Explained | Cumulative % |
| :--- | :---: | :---: | :---: |
| **PC1 (Bioavailability)** | **5.85** | **58.5%** | 58.5% |
| **PC2 (Safety)** | **2.65** | **26.5%** | **85.0%** |
| PC3 | 0.45 | 4.5% | 89.5% |
| PC4 | 0.30 | 3.0% | 92.5% |
| PC5 | 0.25 | 2.5% | 95.0% |
| PC6 | 0.15 | 1.5% | 96.5% |
| PC7 | 0.12 | 1.2% | 97.7% |
| PC8 | 0.10 | 1.0% | 98.7% |
| PC9 | 0.08 | 0.8% | 99.5% |
| PC10 | 0.05 | 0.5% | 100.0% |

* **$z_1$ (PC1):** High variance. Represents **"Bioavailability Bulk."**
* **$z_2$ (PC2):** High variance. Represents **"Biological Safety."**
* **$z_3 \dots z_{10}$:** Near-zero variance. These axes represent redundant information or tiny fluctuations. We take only the first two columns $$V_{reduced} = \begin{pmatrix} 
\vdots & \vdots \\
v_1 &  v_{2} \\
\vdots & \vdots 
\end{pmatrix} \in \mathbb{R}^{10x2}$$ of $V$ and multiply it by our original $1 \times 10$ drug vector $x$ which yields the reduced feature vector.

$$\text{z}_{reduced} = x  V_{reduced}  \in \mathbb{R}^{2}$$

The "collapse" is a conscious choice to simplify the data. We **truncate** the vector $z$ by keeping only the first $k$ dimensions (where $k=2$).

$$z_{\text{reduced}} = [z_1, z_2]$$

By discarding $z_3$ to $z_{10}$, we reduce the dimensionality from **10D to 2D** while retaining the core "story" of the drug. We are left with a 2D signature that tells a chemist everything they need to know about a molecule's potential success.

#### Cave
PCA makes the implicit assumption that "variance" and "meaning" correlate in a meaningful sense. If variance isn't explaining "meaning" or "causation"  PCA can perform badly for predictive models.
